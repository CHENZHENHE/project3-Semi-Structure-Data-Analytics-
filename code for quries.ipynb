1.Show top 10 products (name) sales (quantity x price). 

```python
import pandas as pd
import json

# Create an empty list to store the JSON objects
data_list = []

# Open the JSON file and read it line by line
with open(r'C:\Users\何宸臻1\Desktop\bigdata\project3\sales.json', 'r') as file:
    for line in file:
        # Parse each line as a JSON object and append it to the list
        data_list.append(json.loads(line))

# Create a DataFrame from the list of JSON objects
data = pd.DataFrame(data_list)

# Flatten the 'items' column into separate rows
data = data.explode('items')

# Calculate the total sales (quantity x price) for each item
data['sales'] = data['items'].apply(lambda item: item['quantity'] * float(item['price']['$numberDecimal']))

# Group by item name, sum the total sales, and find the top 10 products
top_10_products = data.groupby(data['items'].apply(lambda x: x['name']))['sales'].sum().nlargest(10)

# Rename the columns
top_10_products = top_10_products.reset_index(name='sales')
top_10_products.rename(columns={'items': 'name'}, inplace=True)

# Display the top 10 products with updated column names
top_10_products

```

2.Show top 3 products (name) sales by store (location). 

```python
# Group by both 'name' and 'storeLocation', sum the total sales, and find the top 3 products for each store
top_3_products_by_store = data.groupby([data['items'].apply(lambda x: x['name']), 'storeLocation'])['sales'].sum().groupby('storeLocation', group_keys=False).nlargest(3)

# Rename the columns
top_3_products_by_store = top_3_products_by_store.reset_index()
top_3_products_by_store.rename(columns={'items': 'name'}, inplace=True)

# Display the top 3 products by name and their sales by store location
top_3_products_by_store

top_3_products_by_store
```

3.Show rankings of each store (location). 

```python
# Group by 'storeLocation' and sum the total sales for each store
store_rankings = data.groupby('storeLocation')['sales'].sum().reset_index()

# Rank the stores based on total sales
store_rankings['rank'] = store_rankings['sales'].rank(ascending=False, method='min')

# Sort the stores by rank
store_rankings = store_rankings.sort_values(by='rank')

# Display the rankings of each store
store_rankings
```

4.Show purchased method by gender table 

```python
import pandas as pd

# Assuming the 'gender' information is in the 'customer' column, you can extract it like this:
data['gender'] = data['customer'].apply(lambda x: x['gender'] if 'gender' in x else None)

# Exclude the 'Phone' purchaseMethod
data = data[data['purchaseMethod'] != 'Phone']

# Pivot the data to create the desired table
purchase_method_gender_table = pd.crosstab(index=data['gender'], columns=data['purchaseMethod'])

# Rename the columns for better clarity
purchase_method_gender_table.rename(columns={'Online': 'Online', 'In store': 'In-Store'}, inplace=True)

# Rename the index for better clarity
purchase_method_gender_table.index = ['Female', 'Male']

# Remove the title for columns
purchase_method_gender_table.columns.name = None

# Display the table
purchase_method_gender_table
```

5.Show monthly total sales 

```python
import pandas as pd

# Load JSON data into a DataFrame
data = pd.read_json(r'C:\Users\何宸臻1\Desktop\bigdata\project3\sales.json', lines=True)

# Custom function to parse the date strings
def custom_date_parser(date_str):
    if '$date' in date_str:
        return pd.to_datetime(date_str['$date'])
    else:
        return None

# Apply the custom date parser to the 'saleDate' column
data['saleDate'] = data['saleDate'].apply(custom_date_parser)

# Define a function to calculate the sales for each row
def calculate_sales(row):
    sales = sum(item['quantity'] * float(item['price']['$numberDecimal']) for item in row['items'])
    return sales

# Apply the sales calculation function to each row
data['sales'] = data.apply(calculate_sales, axis=1)

# Extract month and year from 'saleDate' and create new columns
data['Year'] = data['saleDate'].dt.year
data['Month'] = data['saleDate'].dt.month

# Group by 'Year' and 'Month', and calculate the sum of 'sales'
monthly_sales = data.groupby(['Year', 'Month'])['sales'].sum()

# Display the result
print(monthly_sales)

```
